```python
import pandas as pd

# Load the UK stations data
df_stations = pd.read_csv('/Users/trishapunamiya/Downloads/stations.csv')

# Filter for Scotland only
scotland_stations = df_stations[df_stations['constituentCountry'] == 'scotland'].copy()

print(f"Found {len(scotland_stations)} stations in Scotland")

# Flag major cities
major_cities = ['Edinburgh Waverley', 'Glasgow Central', 'Glasgow Queen Street', 
                'Aberdeen', 'Inverness', 'Dundee', 'Stirling']

scotland_stations['is_major'] = scotland_stations['stationName'].isin(major_cities)

# Check which major cities were found
print("\nMajor cities found:")
print(scotland_stations[scotland_stations['is_major']][['stationName', 'lat', 'long']])

# Save as JSON for your map
scotland_stations.to_json('charts/scotland_train_stations.json', orient='records', indent=2)

print("\n✓ Scotland stations saved with major city flags")
print(f"✓ {scotland_stations['is_major'].sum()} major cities flagged")
print(f"✓ {len(scotland_stations) - scotland_stations['is_major'].sum()} regular stations")
```
